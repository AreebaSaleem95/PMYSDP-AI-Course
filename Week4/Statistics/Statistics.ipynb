- Also Known as Measure of Dispersion.

Agar hum sirf mean dekhta hain tu huma ya nahe pata chalta k data points mean se kitna dor hain, esi liya hum Measure of Dispersion ky help lata hain taka huma ya samjhna ma asani ho k data kitna spread out hy.

# 2.1: Range

- In Descriptive Statistics, range is a simple measure of variability or spread in a dataset.
- It indicates how far the data points are spread out by showing the difference between maximum and minimum values in a dataset

- **Key Points**

- Range gives a quick sense of dispersion: If the range is small, it indicates that the data points are close to each other. A large range suggest that the data points are more spread out.
- Range is sensitive to outliers: Since the range only looks at the two extreme values (Maximum and Minimum), any outlier can significantly increase the range, even if the rest of the data is closely packed.

# 2.2: Variance

- Variance ak statistical measure hy jo dataset ky values k beech ky variability ya dispersion ko quantify krta hy.
- Means that k , ya batata hy k data ky individual values mean say kitni dor hain, or ya dataset ka spread kitna hy.

- Population variance and Sample variance are measure of dispersion that quantify how much data points differ from the mean.

Using numpy

```python
import numpy as np

data = [2,5,6,6,7,5,9,14,10,2,3,4]
df = [2,4,6]
# agr hum log ddof nahe likh tb wo by default sample variance k liya nikalta hy
sample_variance = np.var(data,ddof=1) # ddof -> Delta Degree of Freedom

population_variance = np.var(data,ddof=0) 

sample_variance1 = np.var(df,ddof=1) # ddof -> Delta Degree of Freedom

population_variance1 = np.var(df,ddof=0)
print(sample_variance)
print(population_variance)
print(sample_variance1)
print(population_variance1)
```

Using Pandas

```python
import pandas as pd

data = [2,5,6,6,7,5,9,14,10,2,3,4]

series = pd.Series(data)
sample_variance = series.var()
print(sample_variance)

population_variance = series.var(ddof=0)
print(population_variance)
```

Using scipy

```python
from scipy import stats

data = [2,5,6,6,7,5,9,14,10,2,3,4]

sample_variance = stats.tvar(data)
print(sample_variance)

sample_variance = stats.tvar(data,ddof=0)
print(sample_variance)
```

# 2.3: Standard Deviation

- Standard Deviation ak important measure of dispersion hy statistics ma.
- ya batata hy k apka data points average k ass-pss kitna spread huwa hain.
- Simple terms ma, Standard Deviation sy apko pata chalta hy k apki values average say kitni dor ya pass hain

 Using numpy

```python
import numpy as np

data = [2,5,6,6,7,5,9,14,10,2,3,4]


sample_std = np.std(data,ddof = 1)
print(sample_std)

population_std = np.std(data,ddof=0)
print(population_std)
```

Using pandas

```python
import pandas as pd

data = [2,5,6,6,7,5,9,14,10,2,3,4]

series = pd.Series(data)
sample_std = series.std(ddof=1)
print(sample_std)

population_std = series.std(ddof=0)
print(population_std)
```

Using scipy

```python
from scipy import stats
import math
data = [2,5,6,6,7,5,9,14,10,2,3,4]

sample_std = math.sqrt(stats.tvar(data,ddof=1))
print(sample_std)

population_variance = math.sqrt(stats.tvar(data,ddof=0))
print(population_variance)
```

# 5: Multimodal Distribution

- Multimodal Distribution ak statistical distribution hy jisma data ky 2 say zyada distinct peaks ya modes hoti hain.
- Ya distribution complex datasets ko represent krta hy jahan data multiple distinct groups ya clusters me divide hota hy.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a Figure and axes
fig , axs = plt.subplots(3,2,figsize=(14,12))
fig.suptitle("Histogram with KDE of Various Distributions", fontsize=16)

# Normal Distribution 

data_normal = np.random.normal(loc=0,scale=1,size=1000) # Mean=0, std=1
sns.histplot(data_normal,kde=True,ax=axs[0,0],color="blue",stat="density",bins=30)
axs[0,0].set_title("Normal Distribution")
axs[0,0].grid(True)

data_right_skewed = np.random.exponential(scale=1,size=1000)
data_left_skewed = -np.random.exponential(scale=1,size=1000)

# Right Skewed Histogram
sns.histplot(data_right_skewed,kde=True,ax=axs[0,1],color="red",stat="density",bins=30,label="Right Skewed")


# left Skewed Histogram
sns.histplot(data_left_skewed,kde=True,ax=axs[0,1],color="green",stat="density",bins=30,label="Left Skewed")
axs[0,1].set_title("Skewed Distribution")
axs[0,1].legend()
axs[0,1].grid(True)

# Uniform Distribution
data_uniform = np.random.uniform(low=0,high=1,size=1000)
sns.histplot(data_uniform,kde=True,ax=axs[1,0],color="purple",stat="density",bins=30)
axs[1,0].set_title("Uniform Distribution")
axs[1,0].grid(True)


# Bimodal Distribution
data_bimodal = np.concatenate([np.random.normal(loc=-2,scale=1,size=500),
                               np.random.normal(loc=2,scale=1,size=500)])
sns.histplot(data_bimodal,kde=True,ax=axs[1,1],color="orange",stat="density",bins=30)
axs[1,1].set_title("Bimodal Distribution")
axs[1,1].grid(True)

# Multimodal Distribution
data_Multimodal = np.concatenate([np.random.normal(loc=-6,scale=1.5,size=500),
                               np.random.normal(loc=0,scale=1,size=500),
                               np.random.normal(loc=6,scale=1.5,size=500)])
sns.histplot(data_Multimodal,kde=True,ax=axs[2,0],color="orange",stat="density",bins=30)
axs[2,0].set_title("Multimodal Distribution")
axs[2,0].grid(True)

# Hide Empty subplot
axs[2,1].axis("off")
# adjust layout
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

# Techniques in Inferential Statistics

# 1: Estimate

- Inferential Statistics ma "estimate" ka matlab hota hy k kisi population parameter (jasa population ka mean ya proportion) ko ak sample data k basis per approximate krna.
- jab bhe hum kisi bara group (population) k bara ma kuch conclude krta hain sample data k use se to usse "estimate" kahta hain.

    Types os Estimate in Inferential Statistics:

    - 1: Point Estimate
    - 2: Interval Estimate

# 2: Confidence Interval

- Confidence Interval ak range hy jo estimate krta hy k kisi population parameter ky value kis range k andar fall kr sakti hy, or yeh ak confidence level k sath hota hy.
- Confidence level ya batata hy agar ap multiple sample lain ga or un k confidence intervals calculate karain ga, to usma se kitna intervals population parameters ko cover karain ga.

    For Example:
    - agar ap k sample mean k liya 95% confidence interval 45 se 55 hain, eska matlab hy k apki population mean ka 95% chance hy k ya range (45 se 55) k andar ho ga.

# 3: Hypothesis Testing

- Inferential Statistics ma ak fundamental method hy jo help krta hy k data ky basis per decision lana or conclusion draw krna ma.
- Ya process decide krta hy k sample data ky basis per population k bara ma koi specific claim ya hypothesis true hy k nahe.

    Hypothesis testing hy kya?
    - Hypothesis testing ak systematic method hy jisma num ak hypothesis (assumption) ko test krta hain or decide krta hain k sample data us hypothesis ko support krta hy k nahe.
    - Ya statistical inference ka ak part hy jo k probability theory pe based hota hy.

Hypothesis Testing Machanism:

- 1: Null Hypothesis (H0)
The assumptions you are beginning with

- 2: Alternative Hypothesis (H1)
Opposite of Null Hypothesis

- 3: Experiments
Use Statistical methods (such as t-tests, z-tests, ANOVO, chi-sqaure tests etc) to analyze the data

- 4: Accept the null hypothesis or reject the null hypothesis
It helps researchers and analysts assess the validity of claims, evaluate the effectiveness, interventions and draw meaningful conclusions about populations.

# Tests

p-Value Test:

- p-value Test inferential statistics ma ak key concept hy jo hypothesis testing k process ka essential part hy.

- ya ak measure hy jo batata hy k sample data ko null hypothesis k sath jitna consistent hy.

- Specifically, p-value se ya assess kiya jata hy k null hypothesis ko reject krna k liya apka sample data ky basis per kitna evidence hain.

    p-Value kya hy?

    - p-value ak probability hy jo apko sample data usse |extreme results ana ky probability ko represent krta hy, assuming k null hypothesis sach hy.
    
    - ya batata hy k ap ky observed results (ya or extreme results) ko observe krna kitna unlikely hy agar null hypothesis sach hoti

T Test:

- In Statistics, a t-test is a hypothesis test used to determine if there is a significant difference between the means of two groups.

- It is especially useful when dealing with samll sample sizes or when the population standard deviation is unknown.

```python
import numpy as np
from scipy import stats

# Step 1: Create two random sample datasets
# Suppose these are test scores from two groups (e.g., group A and group B)

group_A = np.random.normal(75,10,30) # Group_A with mean 75, std 10, and 30 samples
group_B = np.random.normal(80,12,30) # Group_B with mean 80, std 12, and 30 samples


# Step 2: Perform an independent t-test

t_statistics, p_value = stats.ttest_ind(group_A,group_B)

# Step 3: Display the p-value and t-statistics

print(f"T-statistics: {t_statistics}")
print(f"p-value: {p_value}")

# Step 4: Interpretation based on p-value
alpha = 0.5
if p_value <= alpha:
    print("reject the null hypothesis (Significant difference between groups)")
else:
    print("Fail to reject the null hypothesis (No Significant difference between groups)")
```

Z-Test:

- Z-Value Test(Z-Test) inferential statistics ka ak method hota hy, jo tab use hota hy jab apko test krna ho k sample k mean population mean k barabar hy k nahe.

- Z-Test ko tab use kiya jata hy jab sample size bara ho (usually n>30n >30n >30) ya jab population ka standard deviation sigma pata ho.

Question:

- A doctor believes the mean to be different.

    - 1: State null and alternative hypothesis
    - 2: At a 95% confidence level, is there enough evidence to reject the null hypothesis 

Ans:

- 1: H0 -> population mean = 168
- 2: H1 -> population mean is not equal to 168

Imagine a company claims that the average weight of their product is 500 grams.
we want to test if this claim is true by collecting a sample of 30 products. the sample mean is 505 grams, and the population standard deviation is 10 grams. 
we will conduct a z-test to see if the difference in the mean is significant or not.

```python
import numpy as np
from scipy import stats

# Step 1: Difine the known parameters

population_mean = 500
sample_mean = 505
population_std = 10
sample_size = 30

# Step 2: Calculate the z-value
z_value = (sample_mean-population_mean)/(population_std/np.sqrt(sample_size))
print(f"Z-Value: {z_value}")

# Step 3: Calculate the p-value (two tailed test)

p_value = 2*(1-stats.norm.cdf(abs(z_value)))
print(f"p-Value: {p_value}")

# Step 4: Set significance Level
alpha = 0.05

# Step 5: Compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis (Significant difference)")
else:
    print("Failed to reject the null hypothesis (no significant difference)")
```
