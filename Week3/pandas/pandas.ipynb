# Filling Missing Values

### 1: Fixed Value Se Fill

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Hina","CR","Monkey"],
    "Marks": [85, np.nan, 78, None,65,0],
    "City": ["Lahore", "Karachi", None, "Islamabad","Patuki","Jungle"]
}

df = pd.DataFrame(data)
df
```

```python
df["Marks"].fillna(10)
```

### 2: Mean Se Fill

```python
# df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].mean())
```

### 3: Median Se Fill

```python
df["Marks"].fillna(df["Marks"].median())
# df["Marks"].fillna(df["Marks"].median(), inplace=True)
```

### 4: Mode Se Fill (Categorical Data)

```python
df["City"].fillna(df["City"].mode()[0], inplace=True)
```

### 5: Forward Fill (Previous Value Copy)

```python
# df.fillna(method="ffill", inplace=True)
df.ffill()
```

### 6: Backward Fill

```python
df.bfill()
```

# Merging (SQL jaisa kaam karta hai)
- pd.merge() use hota hai jab do DataFrames ko common column ki base par combine karna ho.

```python
import pandas as pd
import numpy as np

data1 = {
    "ID":[1,2,3],
    "Name":["Amna ALi","Ammara Akhtar","Abdullah Tahir"]
}

df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
    "ID":[1,5,6],
    "Marks":[95,90,70]
}

df2 = pd.DataFrame(data2)
df2
```

```python
df = pd.merge(df1,df2,on="ID")
```

```python
df
```

# Types of Merging

# 1: Inner Join

- Sirf common IDs rakhta hai.

```python
pd.merge(df1, df2, on="ID", how="inner")
```

# Outer Join

- Dono tables ka sab data rakhta hai.
- Dono tables ki saari unique rows rakhta hai.
- Outer Join Rows = Left + Right − Common

```python
pd.merge(df1, df2, on="ID", how="outer")
```

# Left Join
- Left table ke sab rows rakhta hai.

```python
pd.merge(df1, df2, on="ID", how="left")
```

# Right Join

- Right table ke sab rows rakhta hai.

```python
pd.merge(df1, df2, on="ID", how="right")
```

# Discretization & Binning

# Discretization kya hoti hai?

- 👉 Continuous numbers ko groups (categories) me convert karna.

Matlab:

Age = 23, 45, 67, 12
Isko aise bana dena:

Teen, Young, Adult, Senior

Ye process hi Discretization hai.

# Binning kya hoti hai?

Binning ka matlab:

- 👉 Numbers ko intervals (ranges) me divide karna.

Example:

Agar ages hain:

12, 18, 25, 30, 45, 60

To hum bins bana sakte hain:

0–18
19–30
31–50
51–70

Har number apne bin me chala jayega.

```python
import pandas as pd

ages = [12, 18, 25, 30, 45, 60]
df = pd.DataFrame({"Age": ages})
```

```python
import pandas as pd

Marks = [12, 18, 25, 30, 45, 60,70,75,85,95]
df = pd.DataFrame({"Marks": Marks})
df
```

```python
df["Pass"] = pd.cut(df["Marks"],
                         bins=[0,49,70,89,100],
                         labels=["Fail","C","B","A+"])
```

```python
df
```

```python
import pandas as pd

saman = ["Apple","Mango","Tomato","Oil","Milk","Soap","Lemon","Potato","Dhanya"]

df = pd.DataFrame({"Groceries":saman})
df
```

```python
df["Groceries_Groups"] = pd.cut(df["Groceries"],
                                bins=["Apple","Mango","Tomato","Oil","Milk","Soap","Lemon","Potato","Dhanya"],
                                labels=["Fruits","Daily Usage","Vegetables"] 
)
```

```python
import pandas as pd

saman = ["Apple","Mango","Tomato","Oil","Milk","Soap","Lemon","Potato","Dhanya"]

df = pd.DataFrame({"Groceries": saman})

mapping = {
    "Apple": "Fruits",
    "Mango": "Fruits",
    "Lemon": "Fruits",
    "Tomato": "Vegetables",
    "Potato": "Vegetables",
    "Dhanya": "Vegetables",
    "Oil": "Daily Usage",
    "Milk": "Daily Usage",
    "Soap": "Daily Usage"
}

df["Groceries_Groups"] = df["Groceries"].map(mapping)

print(df)
```

#### Equal Width Binning (pd.cut)

Yahan hum khud ranges define karte hain.

```python
df["Age_Group"] = pd.cut(df["Age"],
                         bins=[0,18,30,50,70],
                         labels=["Teen","Young","Adult","Senior"])
```

```python
df
```

#### Equal Frequency Binning (pd.qcut)

Yahan har group me equal number of data points aate hain.

```python
df["Age_Group"] = pd.qcut(df["Age"], q=3)
```

```python
df
```

Yahan:

- Data ko 3 groups me divide karega

- Har group me almost same count hoga

#### Kab use karte hain?
##### ✔ cut() use karo jab:

- Tumhe specific ranges chahiye

- Jaise grading system:
    - 0–50 = Fail
    - 51–75 = Pass
    - 76–100 = A Grade

##### ✔ qcut() use karo jab:

- Data ko evenly divide karna ho

- Jaise customers ko:
    - Low spenders
    - Medium spenders
    - High spenders

# Pandas Plotting

```python
df.plot()
```
