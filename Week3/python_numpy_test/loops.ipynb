# Q: Reverse a number using while loop.
- Example: 123 → 321

```python
num = input("Enter a number: ")
reversed_num = ""
i = len(num) - 1

while i >= 0:
    reversed_num += num[i]
    i -= 1

print("Reversed number:", int(reversed_num))
```
