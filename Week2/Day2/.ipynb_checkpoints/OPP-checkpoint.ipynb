```python
s1 = Student()
```

```python
class Student:
    name = "Ali"
    age = 20

s1 = Student()

print(s1.name)
print(s1.age)
```

# Instance Variables & Methods

```python
class Student:
    
    def set_data(self, name, age):
        self.name = name
        self.age = age
        
    def show(self):
        print(self.name, self.age)

s1 = Student()
s1.set_data("Ali", 20)
s1.show()
```

# Class Variables & Functions

```python
class Student:
    
    school = "ABC School"   # class variable
    
    def set_data(self, name):
        self.name = name
        
s1 = Student()
s2 = Student()

print(s1.school)
print(s2.school)
```

# Class Function (Class Method)

```python
class Student:
    
    school = "ABC School"
    
    @classmethod
    def change_school(cls, new_name):
        cls.school = new_name

Student.change_school("XYZ School")
print(Student.school)
```

# Constructor

```python
class Student:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
s1 = Student("Ali", 20)

print(s1.name)
print(s1.age)
```

# Destructor

```python
class Test:
    
    def __del__(self):
        print("Object destroyed")

t = Test()
del t
```

# Real-Life Example

```python
class Student:
    
    school = "ABC School"
    
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        
    def show(self):
        print(self.name, self.marks, Student.school)

s1 = Student("Ali", 85)
s2 = Student("Sara", 90)

s1.show()
s2.show()
```
