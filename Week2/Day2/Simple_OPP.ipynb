# Encapsulation

```python
class BankAccount:
    
    def __init__(self, balance):
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
```

```python
b = BankAccount(10)
```

```python
print(b.balance)
```

# Polymorphism

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

d = Dog()
d.sound()
```

```python
class Dog:
    def sound(self):
        return print("Bark")

class Cat:
    def sound(self):
         return print("Meow")

d = Dog()
print(d.sound())
```

# Abstraction

```python
class Car:
    
    def start(self):
        self.__engine_start()
        print("Car started")
        
    def __engine_start(self):
        print("Engine running")

c = Car()
c.start()
```
