# Access Specifiers

# Public

```python
class Test:
    x = 10

obj = Test()
print(obj.x)
```

# Protected (_var)

```python
class Test:
    _y = 20

obj = Test()
print(obj._y)   # allowed but not recommended
```

# Private (__var)

```python
class Test:
    __z = 30

obj = Test()
# print(obj.__z)   ❌ Error
```

# Name Mangling

```python
class Test:
    __z = 30

obj = Test()

print(obj._Test__z)
```
