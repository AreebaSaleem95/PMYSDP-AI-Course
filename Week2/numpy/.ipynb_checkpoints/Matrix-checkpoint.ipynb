# Matrix in NumPy

```python
import numpy as np

A = np.array([[1,2],
              [3,4]])

print(A)
```

# Matrix Addition
Shape same honi chahiye

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

print(A + B)
```

# Matrix Multiplication (Dot Product)
(row × column)

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

print(np.dot(A, B))
```

```python
A @ B # shorthand property
```

# Transpose (Rows ↔ Columns)

```python
A = np.array([[1,2,3],
              [4,5,6]])

print(A.T)
```
