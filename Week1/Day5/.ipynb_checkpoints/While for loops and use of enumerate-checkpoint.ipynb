# *While, for loops and use of enumerate*

# While Loop

```python
👉 Jab tak condition True ho, loop chalta rehta hai
```

```python
while condition:
    code
```

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

# For Loop

```python
👉 Sequence par iterate karta hai (list, string, range, etc.)
```

```python
for i in range(5):
    print(i)
```

```python
fruits = ["apple", "banana", "mango"]

for f in fruits:
    print(f)
```

```python
student = {
    "name":["Ali","Faizan","Abdullah"]
}
for value in student.values():
    print(value)
```

# enumerate() ka Use

```python
👉 Index + value dono deta hai 🔥
```

```python
fruits = ["apple", "banana", "mango"]

for index, value in enumerate(fruits):
    print(index, value)
```

# Nested Loops

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

# List Comprehension

```python
squares = [x for x in range(111)]
print(squares)
```

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

# Iterables

```python
for ch in "HELLO":
    print(ch)
```

# Iterators

```python
nums = [10, 20, 30]

it = iter(nums)

print(next(it))
print(next(it))
print(next(it))
```

```python
student = {
    "name":["Ali","Ammara","Amna","Abdulrehman"],
    "marks":[56,95,95,90]
}
for value in student["name"]:
    print(value)
```

```python
student = {
    "name": ["Ali","Ammara","Amna","Abdulrehman"],
    "marks": [56,95,95,90]
}

for name,marks in zip(student["name"],student["marks"]):
    print(name,":",marks)
```

```python
name= ["Ali","Ammara","Amna","Abdulrehman"]
for value in name:
    print(value)
```

```python
student = {
    "name":["Ali","Ammara","Amna","Abdulrehman"],
    "marks":[56,95,95,90]
}
for value in student.values():
    print(value)
```

```python
# Prints all letters except '45' and '90'

for i in '1-45-78-98-45-75-90'.split("-"):

    if i == '45' or i == '90':
        continue
    print(i)
```

```python
for i in '11 23 24 22 33 44 55 66'.split(" "):

    # break the loop as soon it sees '22'
    # or '55'
    if i == '22' or i == '55':
        break

    print(i)
```
